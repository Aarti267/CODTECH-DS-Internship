```python
import pandas as pd

# Load dataset
df = pd.read_csv("yellow_tripdata_2015-01.csv")

# Display basic info
print("Dataset Info:")
print(df.info())

# Show first 5 rows
print("\nSample Data:")
print(df.head())

# Convert pickup datetime to proper format
df['tpep_pickup_datetime'] = pd.to_datetime(df['tpep_pickup_datetime'])

# Total number of trips
total_trips = len(df)
print("\nTotal Trips:", total_trips)

# Average trip distance
avg_distance = df['trip_distance'].mean()
print("Average Trip Distance:", round(avg_distance, 2))

# Trips per passenger count
trips_by_passenger = df['passenger_count'].value_counts()
print("\nTrips by Passenger Count:")
print(trips_by_passenger)

# Peak pickup hours
df['hour'] = df['tpep_pickup_datetime'].dt.hour
peak_hours = df['hour'].value_counts().sort_index()
print("\nTrips per Hour:")
print(peak_hours)

```
